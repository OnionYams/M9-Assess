```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.preprocessing as pp
from sklearn.cluster import Birch

animals = pd.read_csv('2021-11-ml-09-k-means-animals-dataset.csv', encoding = 'utf-8-sig', decimal=',')

animals = animals.drop(['Region', 'Stage', 'Individual ID', 'Comments'], axis=1)
#animals = animals.drop(['Region', 'Stage', 'Individual ID'], axis=1)
#animals['Comments'] = animals['Comments'].isna().astype(int)

animals.dropna(inplace=True)

scaler = pp.StandardScaler() 
scaled_df = scaler.fit_transform(animals) 
  
# Normalizing the Data 
normalized_df = pp.normalize(scaled_df) 
  
# Converting the numpy array into a pandas DataFrame 
normalized_df = pd.DataFrame(normalized_df) 

print(normalized_df.shape)

from sklearn.metrics import silhouette_score

range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
silhouette_avg = []
X = normalized_df
for num_clusters in range(2,9):
    brc = Birch(threshold=.3,branching_factor=50, n_clusters=num_clusters)
    brc.fit(X)
    cluster_labels = brc.labels_
 
 # silhouette score
    silhouette_avg.append(silhouette_score(X, cluster_labels))
plt.plot(range_n_clusters, silhouette_avg,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Silhouette score') 
plt.title('Silhouette analysis For Optimal k')
plt.show()
```

```python
#https://www.kaggle.com/code/pmrich/clustering-approaches-k-mean-birch-agg/notebook

from sklearn.cluster import KMeans

figure = plt.figure(figsize=(21, 7))
kmean_sil = figure.add_subplot(1,3,1) #kmean silhouette bar chart
birch_sil = figure.add_subplot(1,3,2) #birch chart
kmean_silhouette_scores = [] 
birch_silhouette_scores = [] 
for n_cluster in range(2, 9):
    kmean_silhouette_scores.append( 
        silhouette_score(normalized_df, KMeans(n_clusters = n_cluster).fit_predict(normalized_df))) 
    
    birch_silhouette_scores.append( 
        silhouette_score(normalized_df, Birch(threshold=.3, n_clusters = n_cluster).fit_predict(normalized_df))) 
    
k = [2, 3, 4, 5, 6,7,8] 
kmean_sil.plot(k, kmean_silhouette_scores) 
kmean_sil.set_title('K-MEANS: Number of Clusters vs. Silhouette Score', fontsize = 10)
kmean_sil.set_xlabel('Number of Clusters', fontsize = 20) 
kmean_sil.set_ylabel('Silhouette Score', fontsize = 20) 

birch_sil.plot(k, birch_silhouette_scores) 
birch_sil.set_title('BIRCH: Number of Clusters vs. Silhouette Score', fontsize = 10)
birch_sil.set_xlabel('Number of Clusters', fontsize = 20) 
birch_sil.set_ylabel('Silhouette Score', fontsize = 20)

figure.tight_layout()
```
