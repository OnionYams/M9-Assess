Using the scaled data, plot a series of silhouette scores to determine how good the clustering is

```python
#make plots bigger
plt.rcParams['figure.figsize'] = [7, 7]

range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
silhouette_avg = []
X = scaledanimals
for num_clusters in range(2,9):
    brc = Birch(threshold=.3,branching_factor=50, n_clusters=num_clusters)
    brc.fit(X)
    cluster_labels = brc.labels_
 
 # silhouette score
    silhouette_avg.append(silhouette_score(X, cluster_labels))
plt.plot(range_n_clusters, silhouette_avg,'bx-')
plt.xlabel('Number of clusters') 
plt.ylabel('Silhouette score') 
plt.title('Silhouette analysis for optimal number of clusters')
plt.show()
```
